# Image Data Augmentation

```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
# Display one image
def display_one(a, title1 = "Input Image"):
    plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
img = cv2.imread('garfield.png')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img=cv2.resize(RGB_img, (224, 224))
```

```python
display_one(img)
```

```python
img.shape
```

## Horizontal and Vertical Shift Image Augmentation

```python
# example of horizontal shift image augmentation
data = img_to_array(img)
samples = expand_dims(data, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(width_shift_range=[-50, 50])
it = datagen.flow(samples, batch_size=1)
```

```python
a=1
for i in range(2):
    # generate batch of images
    batch = it.next()
    image = batch[0].astype('uint8')
    display_one(image, title1='Augmented image: '+ str(a))
    a=a+1
# show the figure
plt.show()
```

```python
# example of vertical shift image augmentation
data = img_to_array(img)
samples = expand_dims(data, 0)
# create image data augmentation generator
datagen = ImageDataGenerator(height_shift_range=[-50, 50])
it = datagen.flow(samples, batch_size=1)
```

```python
a=1
for i in range(2):
    # generate batch of images
    batch = it.next()
    image = batch[0].astype('uint8')
    display_one(image, title1='Augmented image: '+ str(a))
    a=a+1
# show the figure
plt.show()
```
